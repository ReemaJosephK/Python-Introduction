```python
#Task 4 - Grade Calculation
#Create a Python class named - Student
class Student:
    #define constructor with default value of grade as None.
    def __init__(self,enrollment_ID, student_name, course_name, age, marks, grade=None):
        # create instance variables: enrollment_ID, student_name, course_name, age, marks, and grade
        self.enrollment_ID = enrollment_ID
        self.student_name  = student_name
        self.course_name = course_name
        self.age = age
        self.marks  = marks
        self.grade = grade

    #age validation method
    def age_validation(self):
        #check the age is a postive integer value greater than zero
        if((isinstance(self.age, int) == False) or (self.age < 0)):
            print("Sorry, Invalid age: ",self.age,". Age should be a positive integer value")
        #check if the age<3 and > 60
        elif(self.age>60 or self.age<3):
            print("Sorry, Invalid age: ",self.age,". Age should be greater than 3 upto 60")
        #else its a valid age
        else:
            print("Sorry, Invalid age: ",self.age,". Age should be greater than 3 upto 60")       

    #mark validation method       
    def validate_and_convert_marks(self):
        #ensure that the entered marks are between 0 and 44
        if(self.marks>=0 and self.marks<=44):
            #convert them into a percentage out of 44 and return the result
            perc = self.marks/44*100
            self.marks =perc
            return True
        #else its a valid age
        else:
            print("Sorry, Invalid marks: ",self.marks,". Please ensure that the entered marks are between 0 and 44") 
            self.marks = None
            return False

    #grade validation method       
    def calculate_grade(self):
        retVal =True
        
        #ensure that the entered marks are between 0 and 44
        if(self.marks<0 or self.marks>100):
            self.grade =None
        elif(self.marks>=0 and self.marks<=20):
            self.grade ="F"
        elif(self.marks>=20.01 and self.marks<=40):
            self.grade ="D"        
        elif(self.marks>=40.01 and self.marks<=60):
            self.grade ="C"
        elif(self.marks>=60.01 and self.marks<=80):
            self.grade ="B"
        elif(self.marks>=80.01 and self.marks<=100):
            self.grade ="A"
        else:
            print("invalid mark")
            retVal= False
            
        return retVal

        
# instantiate object
objStudent = Student(100, "student1", "maths", 30, 44)

# fist convert and validate marks
if(objStudent.validate_and_convert_marks()):
    #then calculate grade if its a valid mark
    if(objStudent.calculate_grade()):
        #print grade
        print("grade: ",objStudent.grade)
```
